## Вопрос №1

Что такое NumPy и зачем он используется в анализе данных?

NumPy — это библиотека для языка программирования Python, которая предоставляет поддержку для работы с многомерными массивами и матрицами, а также включает в себя большое количество математических функций для выполнения операций над этими массивами. NumPy является основой для многих других библиотек, используемых в научных вычислениях и анализе данных, таких как Pandas и Matplotlib.

Основные преимущества:

1. NumPy реализует массивы и матричные операции на низком уровне, что позволяет значительно ускорить вычисления по сравнению с обычными списками Python.

2. Многомерные массивы: NumPy поддерживает многомерные массивы , что позволяет удобно работать с данными в различных формах, например, в виде матриц.

3. Библиотека предоставляет множество встроенных функций для выполнения математических операций, таких как линейная алгебра и статистика.

4. NumPy хорошо совмесим с другими библиотеками для анализа данных и машинного обучения, что делает его важным инструментом в питоне.

5. Использование NumPy позволяет писать более компактный и читаемый код, что упрощает процесс разработки и анализа данных.
В целом, NumPy является важным инструментом для аналитиков данных, ученых и разработчиков, работающих с числовыми данными и требующих высокой производительности и удобства в работе с массивами.

## Вопрос №2

Какие функции и API предоставляет NumPy для работы с массивами и вычисления статистических характеристик? Приведите примеры использования API для расчёта медианы и среднего арифметического.

NumPy имеет множество функций и API для работы с массивами и выполнения различных математических операций, включая вычисление статистических характеристик.

### Создание массивов:

1. numpy.array(): Создает массив из списка или кортежа.

2. numpy.zeros(): Создает массив, заполненный нулями.

3. numpy.ones(): Создает массив, заполненный единицами.

4. numpy.arange(): Создает массив с равномерно распределенными значениями в заданном диапазоне.

### Функции:

1. numpy.mean(): Вычисляет среднее арифметическое.

2. numpy.median(): Вычисляет медиану.

3. numpy.std(): Вычисляет стандартное отклонение.

4. numpy.var(): Вычисляет дисперсию.

## Примеры использования:

Вычисление среднего арифметического:

## Создаем массив

data = np.array([1, 2, 3, 4, 5])

## Вычисляем среднее арифметическое
mean_value = np.mean(data)
print(f"Среднее арифметическое: {mean_value}")

## Вычисление медианы

## Создаем массив
data = np.array([1, 3, 3, 6, 7, 8, 9])

## Вычисляем медиану
median_value = np.median(data)
print(f"Медиана: {median_value}")

Matplotlib.PyPlot предоставляет множество возможностей для создания различных типов графиков, каждый из которых служит своей цели в визуализации данных. Вот некоторые из наиболее распространенных типов графиков:

# Вопрос №5
Что такое описательная статистика и какие статистические характеристики она включает?

Описательная статистика — это раздел статистики, который занимается сбором, анализом и представлением данных с целью их описания и обобщения. Она позволяет исследователям и аналитикам получить представление о характеристиках данных, не прибегая к сложным статистическим методам. Описательная статистика включает в себя несколько ключевых статистических характеристик:

1. Мера центральной тенденции:

   Среднее (Mean): Это сумма всех значений, деленная на их количество. Среднее значение дает представление о "центре" данных.

   Медиана (Median): Это значение, которое делит набор данных на две равные части. Медиана дает представление о "центре" данных.

   Мода (Mode): Это значение, которое встречается наиболее часто в наборе данных. Мода может быть полезна для категориальных данных.

2. Меры разброса:

  Размах (Range): Это разница между максимальным и минимальным значениями в наборе данных. Он показывает, насколько широко распределены данные.

  Дисперсия (Variance): Это среднее значение квадратов отклонений каждого значения от среднего. Дисперсия показывает, насколько данные разбросаны вокруг среднего.

  Стандартное отклонение (Standard Deviation): Это квадратный корень из дисперсии. Стандартное отклонение также показывает разброс данных, на сколько таждый элемент множетсва отличается от центрального.

Описательная статистика является важным инструментом для анализа данных, так как она позволяет быстро и эффективно обобщить информацию, выявить основные тенденции и характеристики, а также подготовить данные для дальнейшего анализа или принятия решений.

# Вопрос №6
Как применяются статистические меры в реальных задачах анализа данных?

Статистические меры играют ключевую роль в анализе данных и применяются в различных реальных задачах для извлечения полезной информации и принятия обоснованных решений. Вот несколько примеров их применения:
Анализ продаж: Компании используют средние значения для оценки общего уровня продаж за определенный период. Например, среднее значение продаж за месяц может помочь определить, насколько успешен продукт. Медиана может быть полезна для выявления типичного уровня продаж, особенно если данные содержат выбросы (например, аномально высокие продажи в определенные дни).

1. Исследования потребительского поведения: Мода может использоваться для определения наиболее популярных товаров или услуг среди клиентов. Например, если в опросе о предпочтениях большинство респондентов выбрали один и тот же продукт, это может указывать на его высокую привлекательность.

2. Финансовый анализ: Стандартное отклонение и дисперсия часто применяются для оценки риска инвестиций. Например, если стандартное отклонение доходности акций высоко, это может указывать на высокую волатильность и риск, что важно для инвесторов при принятии решений.

3. Медицинские исследования: В клинических испытаниях медиана времени до наступления определенного события (например, рецидива болезни) может быть использована для сравнения эффективности различных методов лечения. Квартильный анализ помогает выявить, как распределены результаты среди пациентов.

4. Образование: В образовательных учреждениях описательная статистика может использоваться для анализа успеваемости студентов. Например, средний балл по предмету может помочь определить, насколько хорошо студенты усваивают материал, а интерквартильный размах может показать, насколько сильно различаются результаты среди студентов.

5. Социологические исследования: В социологии описательная статистика помогает анализировать данные опросов. Например, исследователи могут использовать графические представления, такие как гистограммы и коробчатые диаграммы, для визуализации распределения ответов и выявления закономерностей в мнениях респондентов.

6. Маркетинговые исследования: Компании могут использовать описательную статистику для анализа данных о потребителях, чтобы лучше понять их предпочтения и поведение. Например, анализ распределения возрастных групп среди клиентов может помочь в разработке целевых маркетинговых стратегий.

В целом, статистические меры позволяют исследователям и аналитикам быстро обобщать данные, выявлять ключевые тенденции и принимать обоснованные решения на основе фактических данных.

 Снижение среднего чека на 5%: Это может указывать на то, что клиенты стали выбирать более дешевые или менее объемные товары, такие как энергетические батончики, вместо более дорогих соков. Таким образом, увеличение спроса на одну категорию товаров может негативно сказаться на другой.

3. Стратегии для увеличения среднего чека:

  Пакетные предложения: Создание акций, где соки предлагаются в комплекте с другими продуктами, например, с энергетическими батончиками. Это может стимулировать клиентов к покупке большего количества товаров.

  Улучшение качества продукции: Проведение анализа качества соков и внесение необходимых изменений для повышения удовлетворенности клиентов. Это может вернуть клиентов, которые перестали покупать соки из-за низкого качества.

  Введение новых вкусов и форматов: Разработка новых вкусов соков или упаковок, которые могут привлечь внимание клиентов и увеличить интерес к продукту.

  Программы лояльности: Введение программ лояльности, которые будут поощрять повторные покупки и увеличивать средний чек. Например, скидки на следующий заказ при покупке определенного количества товаров.

4. Сохранение лояльности существующих клиентов:

  Регулярное получение обратной связи от клиентов через опросы и отзывы. Это поможет понять их потребности и ожидания, а также вовлечь их в процесс принятия решений.

  Сохранение ценовой политики: Если возможно, сохранить цены на популярные товары, чтобы не отпугнуть существующих клиентов.

  Персональные предложения: Использование данных о покупках для создания персональных предложений, которые будут интересны конкретным клиентам.

  Поддержка качества обслуживания: Обеспечение высокого уровня обслуживания клиентов, чтобы они чувствовали себя ценными и важными для компании.

Для улучшения ситуации необходимо сосредоточиться на повышении качества соков, разработке новых стратегий продаж и поддержании лояльности клиентов.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
months = data['Месяц']
pm25_levels = data['Уровень PM2.5']
pm10_levels = data['Уровень PM10']
```

```python
plt.figure(figsize=(10, 5))
plt.plot(months, pm25_levels, color='blue', marker='o', label='PM2.5')
plt.plot(months, pm10_levels, color='orange', marker='o', label='PM10')
plt.title('Изменение уровней PM2.5 и PM10 по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Уровень (мкг/м³)')
plt.xticks(rotation=90)
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(months, pm10_levels, color='orange')
plt.title('Распределение уровней PM10 по месяцам')
plt.xlabel('Месяцы')
plt.ylabel('Уровень PM10 (мкг/м³)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
average_pm25 = (pm25_levels).mean()
above_average = sum(level > average_pm25 for level in pm25_levels)
below_average = len(pm25_levels) - above_average
```

```python
labels = ['Выше среднего', 'Ниже среднего']
sizes = [above_average, below_average]
```

```python
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Доля месяцев с уровнем PM2.5 выше и ниже среднего')
plt.axis('equal')  # Чтобы круг был кругом
plt.show()
```
